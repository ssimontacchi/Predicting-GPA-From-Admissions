## GPA Trends by Year
#### This notebook takes 10 years of academic data and generates graphs to visualize GPA trends over those years.
*Data has been altered for the purposes of confidentitality.

```python
data = pd.read_excel('CumGPAsByYear.xlsx')
summary_stats = data.describe()
gpa_by_year = summary_stats.drop(['count','mean', 'std','min','max'])
gpa_by_year
```

```python
output_notebook()

p = figure(title="Cumulative GPAs By Year",plot_width=400, plot_height=400)
p.xaxis.axis_label = 'Year Graduated'
p.yaxis.axis_label = 'GPA'

leg = ["75%","50%","25%"]
col = ["green", "navy", "firebrick"]
counter = 0
years = [2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017]
for i in [gpa_by_year.values.tolist()[2], 
          gpa_by_year.values.tolist()[1], 
          gpa_by_year.values.tolist()[0]]:
    p.line(x=years, y=i, legend = leg[counter], color = col[counter])
    counter += 1


p.legend.location = "bottom_right"

export_png(p, filename="results/"+title+".png")
```
